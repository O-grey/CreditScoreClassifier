```python
# Define bounds for acceptable age values
lower_bound = 0  # Minimum logical age
upper_bound = 100  # Maximum logical age

# Remove rows where 'Age' is outside these bounds
df1 = df[(df['Age'] >= lower_bound) & (df['Age'] <= upper_bound)]

# Verify the cleaned dataset
df1['Age'].describe(),df1.shape
```

```python
## Creating General Method to remove outliers for all numeric columns:
def remove_outliers(data,method='remove'):
  # Select numeric columns
  num_columns=data.select_dtypes(exclude='O').columns

  for column in num_columns:
     # Calculate quartiles and IQR
    minimum,Q1,median,Q3,maximum=np.quantile(data[column],[0,0.25,0.50,0.75,1.0])
    IQR=Q3-Q1
    lower_bound=Q1-1.5*(IQR)
    upper_bound=Q3+1.5*(IQR)

    if method == 'remove':
            # Remove rows with outliers
            filtered_data = data[(data[column] >= lower_bound) & (data[column] <= upper_bound)]

            # Check if filtering resulted in an empty DataFrame
            if filtered_data.empty:
                print(f"Warning: All rows in column '{column}' are considered outliers and have been removed.")
                continue  # Skip further processing for this column if it's empty
            data=filtered_data

    elif method == 'cap':
            # Cap outliers at thresholds
            data[column] = data[column].clip(lower=lower_bound, upper=upper_bound)

  return data
```

```python
df1_removed=remove_outliers(df1,'remove')
df1_capped=remove_outliers(df1,'cap')
```

```python
def convert_to_months(time_string):
    if pd.isna(time_string):  # Check for missing values
        return None
    parts = time_string.split()
    years = int(parts[0])
    months = int(parts[3])
    return (years * 12) + months

# Apply the function to the Credit_History_Age column
data['Credit_History_Age'] = data['Credit_History_Age'].apply(convert_to_months)

data.head(20)
```

```python
import re
def removing_underscores(string):
    if pd.isna(string):  # Check for missing values
      return None
    else:
      return re.sub(r'[\d\-\,\_\(\)\s]', '', string)


# Apply the function to the Credit_History_Age column
data['Payment_Behaviour'] = data['Payment_Behaviour'].apply(removing_underscores)
data['Type_of_Loan'] = data['Type_of_Loan'].apply(removing_underscores)
data['Occupation'] = data['Occupation'].apply(removing_underscores)

data.head(20)
```

```python
for feature in cat_features_with_na:
  mode_selected= data[data[feature].notna()][feature].mode()[0]
  data[feature].fillna(mode_selected,inplace=True)
  print("Mode selected for ",feature, "is---" ,mode_selected)
```
